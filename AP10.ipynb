```python
from selenium import webdriver  # کد خودم
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
import time

options = Options()
options.add_argument("--start-maximized")

driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=options
)

driver.get("https://mftplus.com/calendar")
time.sleep(10)

search_box = driver.find_element(
    By.XPATH,
    "/html/body/div/div[2]/div[2]/div[2]/div/div[1]/div/input"
)

search_box.send_keys("پایتون")
search_box.send_keys(Keys.RETURN)
time.sleep(5)

cards = driver.find_elements(By.CLASS_NAME, "cards")
classes = []

for card in cards:
    c = card.find_element(By.CSS_SELECTOR, "div:nth-child(1)")
    
    class_data = {
        "register_item": c.get_attribute('data-class-register-item'),
        "class_time": c.get_attribute('data-class-time'),
        "class_number": c.get_attribute('data-class-number'),
        "lesson_id": c.get_attribute('data-lesson-id'),
        "class_start": c.get_attribute('data-class-start'),
        "class_end": c.get_attribute('data-class-end'),
        "price": card.find_element(
            By.CSS_SELECTOR,
            "div.data div.cost div.price div"
        ).text
    }

    classes.append(class_data)

driver.quit()
print(classes)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager

from jdatetime import datetime as jdatetime
from datetime import datetime as gdatetime


# ---------- Driver Setup ----------
options = Options()
options.add_argument("--start-maximized")

driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=options
)

wait = WebDriverWait(driver, 15)

# ---------- Open Page ----------
driver.get("https://mftplus.com/calendar")

# ---------- Search ----------
search_box = wait.until(
    EC.presence_of_element_located(
        (By.XPATH, "/html/body/div/div[2]/div[2]/div[2]/div/div[1]/div/input")
    )
)

search_box.send_keys("پایتون")
search_box.send_keys(Keys.RETURN)

# ---------- Wait for Cards ----------
cards = wait.until(
    EC.presence_of_all_elements_located((By.CLASS_NAME, "cards"))
)

# ---------- Extract & Convert ----------
classes = []

for card in cards:
    c = card.find_element(By.CSS_SELECTOR, "div:nth-child(1)")

    # تاریخ‌های خام از data-attribute
    raw_start = c.get_attribute("data-class-start")
    raw_end = c.get_attribute("data-class-end")

    # تبدیل تاریخ شمسی
    start_jdate = jdatetime.strptime(raw_start, "%A %d %B %Y")
    end_jdate = jdatetime.strptime(raw_end, "%A %d %B %Y")

    # تبدیل به میلادی
    start_gdate = start_jdate.togregorian()
    end_gdate = end_jdate.togregorian()

    try:
        price = card.find_element(
            By.CSS_SELECTOR,
            "div.data div.cost div.price div"
        ).text
    except:
        price = None

    class_data = {
        "lesson_id": c.get_attribute("data-lesson-id"),
        "class_number": c.get_attribute("data-class-number"),

        # شمسی
        "start_jalali": start_jdate.strftime("%Y-%m-%d"),
        "end_jalali": end_jdate.strftime("%Y-%m-%d"),

        # میلادی (مناسب DB / API)
        "start_gregorian": start_gdate.strftime("%Y-%m-%d"),
        "end_gregorian": end_gdate.strftime("%Y-%m-%d"),

        "price": price
    }

    classes.append(class_data)

driver.quit()

# ---------- Result ----------
for item in classes:
    print(item)
```
