```python
import json

def read_all():
    try:
        with open('data4.txt', 'r', encoding='utf-8') as fpr:
            return json.load(fpr)
    except FileNotFoundError:
        return {}

def read_filej(id):
    data = read_all()
    return data.get(id)

def write_filej(data):
    with open('data4.txt', 'w', encoding='utf-8') as fp:
        json.dump(data, fp, ensure_ascii=False, indent=4)


class MyAddress:
    city = "Tehran"

    def __init__(self, street, alley, postal_code=None, id=None):
        self.street = street
        self.alley = alley
        self.postal_code = postal_code
        self.id = id

    def __str__(self):
        return f"{self.street}, {self.alley}, {self.postal_code} (ID={self.id})"

    @classmethod
    def get_last_id(cls):
        data = read_all()
        if not data:
            return 0
        return max(int(k) for k in data.keys())

    def save(self):
        data = read_all()
        
        last_id = MyAddress.get_last_id()
        new_id = str(last_id + 1)

        self.id = new_id

        data[new_id] = {
            "street": self.street,
            "alley": self.alley,
            "postal_code": self.postal_code
        }

        write_filej(data)

    @classmethod
    def load(cls, id):
        obj = read_filej(id)
        if obj is None:
            return None
        return cls(id=id, **obj)



# -----------------------------
# باز سازی تابع لود و سیو از روی فایل جیسون 
# -----------------------------

x = MyAddress.load("2")
print(type(x))
print(x)


```
