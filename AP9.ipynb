```python
from fastapi import FastAPI, Depends
from pydantic import BaseModel
from sqlalchemy import create_engine, Integer, String
from sqlalchemy.orm import sessionmaker, DeclarativeBase, Mapped, mapped_column, Session
from sqlalchemy.types import JSON
import os, nest_asyncio

nest_asyncio.apply()

receiver_app = FastAPI()

# ---- schemas ----
class StudentIn(BaseModel):
    name: str
    id: int
    course: dict[str, int]

class StudentOut(StudentIn):
    db_pk: int

# ---- DB (absolute path to avoid "where is my db?" issue) ----
DB_PATH = os.path.abspath("students.db")
DATABASE_URL = f"sqlite:///{DB_PATH}"
print("âœ… Receiver DB file:", DB_PATH)

engine = create_engine(DATABASE_URL, connect_args={"check_same_thread": False})
SessionLocal = sessionmaker(bind=engine, autoflush=False, autocommit=False)

class Base(DeclarativeBase):
    pass

class StudentDB(Base):
    __tablename__ = "students"
    db_pk: Mapped[int] = mapped_column(Integer, primary_key=True, index=True, autoincrement=True)
    name: Mapped[str] = mapped_column(String(100), nullable=False)
    student_id: Mapped[int] = mapped_column(Integer, nullable=False, index=True)
    course: Mapped[dict] = mapped_column(JSON, nullable=False)

Base.metadata.create_all(bind=engine)

def get_db():
    db = SessionLocal()
    try:
        yield db
    finally:
        db.close()

@receiver_app.get("/")
def root():
    return {"status": "receiver ok"}

@receiver_app.post("/students", response_model=StudentOut)
def receive_student(payload: StudentIn, db: Session = Depends(get_db)):
    row = StudentDB(name=payload.name, student_id=payload.id, course=payload.course)
    db.add(row)
    db.commit()
    db.refresh(row)
    return StudentOut(db_pk=row.db_pk, name=row.name, id=row.student_id, course=row.course)

@receiver_app.get("/students")
def list_students(db: Session = Depends(get_db)):
    rows = db.query(StudentDB).order_by(StudentDB.db_pk.desc()).all()
    return [{"db_pk": r.db_pk, "name": r.name, "id": r.student_id, "course": r.course} for r in rows]
```
